## Key Relationship Stats

## Timeline of Key Events

```python
events = {
    "First Date": "2024-01-22",
    "First 'I Love You'": "2024-02-14",
    "First Trip Together": "2024-04-27",
    "First Wedding Attended": "2024-05-11",
    "Anniversary": "2025-01-22","6-Month Photoshoot": "2024-08-16",
}



# Convert dates to pandas datetime
dates = pd.to_datetime(list(events.values()))
labels = list(events.keys())

# Alternating label positions (above and below the timeline)
positions = [1.02 if i % 2 == 0 else 0.98 for i in range(len(labels))]
alignments = ['top' if i % 2 == 0 else 'bottom' for i in range(len(labels))]

# Create the timeline plot
plt.figure(figsize=(10, 3))
plt.hlines(1, dates.min(), dates.max(), color='gray', alpha=0.6, linestyles="dashed")  # Base timeline
plt.scatter(dates, [1]*len(dates), color="red", s=100, zorder=5)  # Timeline points

# Add labels and vertical lines for better clarity
for i, (txt, pos, align) in enumerate(zip(labels, positions, alignments)):
    plt.text(dates[i], pos, txt, ha="center", fontsize=10, verticalalignment=align)
    plt.vlines(dates[i], 0.95, 1.05, color='gray', linestyles="dotted")

# Formatting the plot
plt.title("Our Relationship Timeline ‚ù§Ô∏è", fontsize=14)
plt.yticks([])
plt.xticks(rotation=30)  # Rotate x-axis labels
plt.tight_layout()
plt.show()
```

## Favourite Date Spots Analysis
